# P1

Por demostrar que $val(f)=f(\delta^{+}(s))-f(\delta^{-}(s))=\sum_{P\in P_{s,t}} \lambda_{P}$:

$$val(f)=\sum_{e\in \delta^{+}(s)}f(e)-\sum_{e\in \delta^{-}(s)}f(e)$$
$$val(f)=\sum_{e\in \delta^{+}(s)}\sum_{P\in P_{s,t}}\lambda_{P}\chi^{P}(e)+\sum_{e\in \delta^{+}(s)}\sum_{C\in _{s,t}}\lambda_{P}\chi^{C}(e)-\sum_{e\in \delta^{-}(s)}\sum_{P\in P_{s,t}}\lambda_{P}\chi^{P}(e)-\sum_{e\in \delta^{-}(s)}\sum_{C\in C_{s,t}}\lambda_{C}\chi^{C}(e)$$
$$ val(f)=\sum_{P\in P_{s,t}}\sum_{e\in \delta^{+}(s)}\lambda_{P}\chi^{P}(e)+\sum_{C\in _{s,t}}\sum_{e\in \delta^{+}(s)}\lambda_{P}\chi^{C}(e)-\sum_{P\in C_{s,t}}\sum_{e\in \delta^{-}(s)}\lambda_{P}\chi^{P}(e)-\sum_{C\in C_{s,t}}\sum_{e\in \delta^{-}(s)}\lambda_{C}\chi^{C}(e)   $$

Luego 

El primer termino, como son caminos que empiezan en $s$, siempre el termino$ \sum_{e\in \delta^{+}(s)}\lambda_{P}\chi^{P}(e)$, valdrá $\lambda_{P}$, ya que solo una arista será saliente de $s$, también el termino $\sum_{C\in C_{s,t}}\sum_{e\in \delta^{-}(s)}\lambda_{C}\chi^{C}(e) $, valdrá 0, ya que no hay aristas entrantes en un camino saliente desde $s$, luego los términos $\sum_{C\in _{s,t}}\sum_{e\in \delta^{+}(s)}\lambda_{P}\chi^{C}(e)- \sum_{C\in C_{s,t}}\sum_{e\in \delta^{-}(s)}\lambda_{C}\chi^{C}(e)   $, serán 0, ya que por cada ciclo,  si es que exise una arista entrante, habrá una saliente, y se cancelaran entre si, si no existe esta arista, la indicatriz solo será 0.


Por demostrar que $c(f)\geq\sum_{P\in P_{s,t}} \lambda_{P}c(\chi^{P})$:
    $c(f)=\sum_{e\in E}c(e)f(e)\geq \sum_{e\in E}c(e)\sum_{P\in P_{s,t}}\lambda_{P}\chi^{P}(e)$, donde se toma la cota  ya que todos los términos de los ciclos son positivos.
    
 $   \sum_{e\in E}\sum_{P\in P_{s,t}}\lambda_{P}\chi^{P}(e)c(e)= \sum_{P\in P_{s,t}}\sum_{e\in E}\lambda_{P}\chi^{P}(e)c(e)=\sum_{P\in P_{s,t}}\lambda_{P} c(\chi^{P})$
 
 

Para responder la segunda pregunta, tomemos un $f$ flujo optimo
    Por el teorema de descomposición de flujo $f=\sum_{P\in P_{s,t}}\lambda_{P}\chi^{P}+ \sum_{C\in C_{s,t}} \lambda_{C}\chi^{C}$, por lo demostrado anteriormente parece razonable tomar el, flujo $f'=\sum_{P\in P_{s,t}}\lambda_{P}\chi^{P}$, que toma solo las partes de los caminos, y supongamos que es factible, veamos que es optimo, se tiene, ya que $f$ es optimo y el costo de $f'$ es menor o igual que el de $f$.




Sigue siendo factible, ya que al quitarle los ciclos  el flujo seguira siendo 0 en los vertices distintos de la fuente, ademas el valor es el mismo por lo demostrado anteriormente.

# P2

Notemos que el problema a resolver corresponde a 
$$(P) \max_{{P\in P_{s,t}}} z -y(p)-c(P)$$
Como el $z$ es constante, resolver este problema es equivalente a resolver
$$(P') \max_{{P\in P_{s,t}}}  -y(p)-c(P)$$
Que es equivalente a 
$$(P'') \max_{{P\in P_{s,t}}}  -(y(p)+c(P))$$
Que tambien es equivalente a 

$$(P'') \min_{{P\in P_{s,t}}}  (y(p)+c(P))$$


Por otro lado, encontrar el camino de costo minimo, cuando la funcion de largos esta dada por $l(e)=c_{e}+y_{e}$
corresponde a 
$$ (P''') \min_{P\in P_{s,t}} \sum_{e\in P} c_{e} + y_{e}$$
Que corresponde justamente a
$$(P'''') \min_{P\in P_{s,t}} c(P) +y(P)$$.


COmo $(P)$ es equivalente a $ (P''')$, que es equivalente a $(P'''')$, se concluye que resolver los problemas $(P)$ y $(P'''')$ son equivalentes.
